```python
# Data for the comparison
comparison_data_3_clusters = {
    "Algorithm": ["NSForest", "NSAUC-ROC", "NSmRMR"],
    "f_score_no_bs": [0.9505, 0.9534, 0.9531],
    "f_score_bs": [0.9531, 0.9534, 0.9531],
    "PPV_no_bs": [0.9774, 0.9808, 0.9786],
    "PPV_bs": [0.9779, 0.9808, 0.9786]
}

comparison_data_14_clusters = {
    "Algorithm": ["NSForest", "NSAUC-ROC", "NSmRMR"],
    "f_score_no_bs": [0.6798, 0.6591, 0.6950],
    "f_score_bs": [0.6630, 0.6569, 0.6533],
    "PPV_no_bs": [0.7398, 0.7165, 0.7383],
    "PPV_bs": [0.7350, 0.7159, 0.7210]
}

comparison_data_14_clusters_split = {
    "Algorithm": ["NSForest", "NSAUC-ROC", "NSmRMR"],
    "f_score_no_bs": [0.7354, 0.6820, 0.7244],
    "f_score_bs": [0.6731, 0.6705, 0.6652],
    "PPV_no_bs": [0.7862, 0.7348, 0.7772],
    "PPV_bs": [0.7380, 0.7402, 0.7331]
}

# Ensure the directory exists
output_dir = '/home/a/Escritorio/TFM/src/NSForest/outputs_experimentation/graficos/'
os.makedirs(output_dir, exist_ok=True)

# Define pastel colors
colors = {
    "f_score_no_bs": "#add8e6",
    "f_score_bs": "#448aff",
    "PPV_no_bs": "#ffc966",
    "PPV_bs": "#ff5722"
}

# Function to create and save the comparison plots
def create_comparison_plot(data, title, filename):
    df = pd.DataFrame(data)
    n_algorithms = len(df['Algorithm'])
    width = 0.2  # the width of the bars
    inner_gap = width * 2  # gap between f_score and PPV bars for the same algorithm
    outer_gap = width * 3  # gap between different algorithms

    x = np.arange(n_algorithms) * (2 * inner_gap + outer_gap)  # the label locations with space for gaps

    fig, ax = plt.subplots(figsize=(14, 8))

    # Use a clean and minimalistic theme
    plt.style.use('seaborn-whitegrid')
    plt.rcParams.update({'font.size': 14, 'font.family': 'serif'})

    # Bars for f_score
    rects1 = ax.bar(x - inner_gap / 2 - width, df['f_score_no_bs'], width, label='f_score (no BS)', color=colors["f_score_no_bs"], edgecolor='black')
    rects2 = ax.bar(x - inner_gap / 2, df['f_score_bs'], width, label='f_score (BS)', color=colors["f_score_bs"], edgecolor='black')

    # Bars for PPV
    rects3 = ax.bar(x + inner_gap / 2, df['PPV_no_bs'], width, label='PPV (no BS)', color=colors["PPV_no_bs"], edgecolor='black')
    rects4 = ax.bar(x + inner_gap / 2 + width, df['PPV_bs'], width, label='PPV (BS)', color=colors["PPV_bs"], edgecolor='black')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_xlabel('Algorithm', fontsize=22, fontweight='bold')
    ax.set_ylabel('Score', fontsize=22, fontweight='bold')
    ax.set_title(f'{title}', fontsize=28, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(df['Algorithm'], fontsize=16, fontweight='bold')

    # Place the legend outside the plot
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2, fontsize=24)

    # Add grid lines
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)
    ax.set_axisbelow(True)

    # Set y-axis limits between 0 and 1
    ax.set_ylim(0, 1)

    # Improve layout and spacing
    fig.tight_layout()
    plt.savefig(os.path.join(output_dir, f'{filename}.png'), bbox_inches='tight', dpi=300)  # Increase resolution for better quality
    plt.close()

# Creating comparison plots
create_comparison_plot(comparison_data_3_clusters, "Comparativa 3 Clusters", "comparativa_3_clusters")
create_comparison_plot(comparison_data_14_clusters, "Comparativa de 14 Clusters", "comparativa_14_clusters")
create_comparison_plot(comparison_data_14_clusters_split, "Comparativa de 14 Clusters con Split Train/Test", "comparativa_14_clusters_split")
```

### Gráficos para Comparativa de Algoritmos sin Preselección de Genes

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Data for the comparison without preselection (5.1.5)
comparison_data_3_clusters_no_preselection = {
    "Algorithm": ["NSForest", "NSAUC-ROC", "NSmRMR"],
    "f_score": [0.9505, 0.9534, 0.9531],
    "PPV": [0.9774, 0.9808, 0.9786]
}

comparison_data_14_clusters_no_preselection = {
    "Algorithm": ["NSForest", "NSAUC-ROC", "NSmRMR"],
    "f_score": [0.6798, 0.6591, 0.6950],
    "PPV": [0.7398, 0.7165, 0.7383]
}

comparison_data_14_clusters_split_no_preselection = {
    "Algorithm": ["NSForest", "NSAUC-ROC", "NSmRMR"],
    "f_score": [0.7354, 0.6820, 0.7244],
    "PPV": [0.7862, 0.7348, 0.7772]
}

# Ensure the directory exists
output_dir = '/home/a/Escritorio/TFM/src/NSForest/outputs_experimentation/graficos/'
os.makedirs(output_dir, exist_ok=True)

# Pastel colors for each algorithm
colors = {
    "NSForest": "#add8e6",  # Light blue
    "NSAUC-ROC": "#ffd700",  # Light yellow
    "NSmRMR": "#98fb98"  # Light green
}

# Function to create and save the comparison plots
def create_comparison_plot(data, title, filename):
    algorithms = data["Algorithm"]
    f_scores = data["f_score"]
    ppvs = data["PPV"]
    
    x = np.arange(2)  # Two groups: f_score and PPV
    width = 0.2  # the width of the bars

    fig, ax = plt.subplots(figsize=(12, 8))

    # Bar positions
    x_f_score = x[0]  # Position for f_score group
    x_ppv = x[1]  # Position for PPV group
    
    # Adjust positions for each algorithm
    positions_f_score = [x_f_score - width, x_f_score, x_f_score + width]
    positions_ppv = [x_ppv - width, x_ppv, x_ppv + width]

    # Plotting the bars with colors and edgecolor
    bars1 = []
    bars2 = []

    for i, algorithm in enumerate(algorithms):
        bars1.append(ax.bar(positions_f_score[i], f_scores[i], width, label=f'{algorithm}', color=colors[algorithm], edgecolor='black'))
        bars2.append(ax.bar(positions_ppv[i], ppvs[i], width, color=colors[algorithm], edgecolor='black'))

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_xlabel('Metrics', fontsize=22, fontweight='bold')
    ax.set_ylabel('Score', fontsize=22, fontweight='bold')
    ax.set_title(f'{title}', fontsize=28, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(['f_score', 'PPV'], fontsize=16, fontweight='bold')

    # Place the legend outside the plot
    handles = [bars1[0], bars1[1], bars1[2]]
    labels = ['NSForest', 'NSAUC-ROC', 'NSmRMR']
    ax.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, fontsize=20)

    # Add values above bars and label algorithms
    def add_values(bars, values):
        for bar_group, value in zip(bars, values):
            bar = bar_group[0]
            height = bar.get_height()
            ax.annotate(f'{round(value, 4)}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=10, fontweight='bold')

    add_values(bars1, f_scores)
    add_values(bars2, ppvs)

    # Set y-axis limits between 0.5 and 1 for the specific plot
    if filename == "comparison_3_clusters_no_preselection.png":
        ax.set_ylim(0.9, 1)
    else:
        ax.set_ylim(0, 1)

    # Add a light grid
    ax.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

    fig.tight_layout()

    plt.savefig(os.path.join(output_dir, filename))
    plt.close()

# Creating comparison plots
create_comparison_plot(comparison_data_3_clusters_no_preselection, "Comparativa 3 Clusters", "comparison_3_clusters_no_preselection.png")
create_comparison_plot(comparison_data_14_clusters_no_preselection, "Comparativa 14 Clusters", "comparison_14_clusters_no_preselection.png")
create_comparison_plot(comparison_data_14_clusters_split_no_preselection, "Comparativa 14 Clusters with Split Train/Test", "comparison_14_clusters_split_no_preselection.png")
```
